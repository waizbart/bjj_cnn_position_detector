## Constantes

```python
dataset_images_path = 'D:\jiu_dataset\images'
dataset_annotations_path = '../assets/annotations.json'

dataset_size = 120278
```

## Testing random values

```python
import random
import json
import matplotlib.pyplot as plt
from utils.draw import draw_image_with_keypoints

with open(dataset_annotations_path, 'r') as f:
    annotations = json.load(f)

for _ in range(10):
    random_n = random.randint(0, dataset_size)

    random_annotation = annotations[random_n]
    
    print(f'Random annotation: {random_annotation}')
    
    dataset_image_path = f'{dataset_images_path}/{random_annotation["image"]}.jpg'
    
    plt.title(random_annotation['position'])
    
    keypoints = []
    
    pose1 = random_annotation.get('pose1')
    pose2 = random_annotation.get('pose2')
    
    if pose1:
        keypoints.append(pose1)
        
    if pose2:
        keypoints.append(pose2)
        
    draw_image_with_keypoints(dataset_image_path, keypoints)
```
